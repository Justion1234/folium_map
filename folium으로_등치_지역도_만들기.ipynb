```python
import folium
import pandas as pd    

geo_json = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

raw_data = [ 
    ['강동구', 0.9912224670671045],
    ['송파구', 0.8203005553596422],
    ['강남구', 0.5236343767243283],
    ['서초구', 0.7718434066818908],
    ['관악구', 0.2671968592866498],
    ['동작구', 0.647485262482158],
    ['영등포구', 0.700692780238187],
    ['금천구', 0.1163259091582638],
    ['구로구', 0.24656487938230498],
    ['강서구', 0.9770965752946363],
    ['양천구', 0.6186435571604079],
    ['마포구', 0.2817506680903893],
    ['서대문구', 0.6190094316405799],
    ['은평구', 0.6367052955383669],
    ['노원구', 0.6512900741098098],
    ['도봉구', 0.16751070816570457],
    ['강북구', 0.8406888155269459],
    ['성북구', 0.15584205835364529],
    ['중랑구', 0.66268679705452016],
    ['동대문구', 0.7288276181704965],
    ['광진구', 0.012988149076251232],
    ['성동구', 0.17540892673418795],
    ['용산구', 0.39018545494125245],
    ['중구', 0.10321901374884213],
    ['종로구', 0.10048909676269702],
 ]   
 

data = pd.DataFrame(raw_data, columns=['name', 'value']) #name : 구 이름, value : 옆에 값

m = folium.Map(
    location=[37.566345, 126.977893],
    tiles='Stamen Terrain',                 
)

#등치 지역도 추가하는 코드
folium.Choropleth( 
    geo_data=geo_json, #등치지역도에 필요한 데이터를(지역별의 모양 데이터) 넣기 url을 넣어도 실행 가능 geo_data = geo_json이라고 해놨는데 위에 geo_json 보면 아까전에 github url 가져온 것
    name = 'choropleth' ,
    data=data, #출력할 값(숫자)
    columns=['name', 'value'],

    key_on = 'feature.properties.name' , #입력했던 geo data안에서 지역의 이름이 어디에 저장되어있는지 표현한거
    fill_color='YlGn' , #YlGn의 의미는 숫자가 작을 때는 노란색 숫자가 크면 클수록 초록색

    fill_opacity=0.7, #색칠했을 때 1이면 완벽하게 색칠한다는 뜻 0.7이면 조금더 투명하게 
    #opacity = 1, 1이면 완전히 불투명
    #0이면 완전 투명

    line_opacity=0.2, #구분선의 투명도

).add_to(m) #add_to라는 메소드를 이용해서 m이라는 지도에 이 등치 지역도 추가

m
```

key_on = 'feature.properties.name'   <- 위에 이렇게 적은 이유를 확인해보기 위해서는 구글에 json preetify를 검색하면 그 위에 바로 나오는 url을 클릭을 하게 되면 https://jsonformatter.curiousconcept.com/이 url로 올텐데
